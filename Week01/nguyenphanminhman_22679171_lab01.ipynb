1. Viết chương trình nhập vào số nguyên n (n>0). In ra các số nguyên tố từ 1..n


````python
import math;

def soNguyenTo(n):
    if (n < 2):
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0: 
            return False
    return True

data = int(input("Nhap n (n>0)"))

for i in range(1, data+1):
    if (soNguyenTo(i)):
        print(i, end=" ")
````

````python

def soDao(n):
    nSoDao = 0
    
    while(n != 0):

        nRem = n % 10

        nSoDao = (nSoDao * 10) + nRem

        n = n / 10
    return n

def soDoiXung(n):
    if (n == soDao(n)):
        return True
    else:   
        return False
    
n = int(input("Nhap so nguyen duong n(n>0):"))
if soDoiXung(n):
    print("true")
else:
    print("false")



````

10. Viết hàm có tham số đầu vào là một list L có các phần tử là chuỗi. Hãy tạo ra một dictionary D mã hóa, với mỗi một phần tử trong L được mã hóa thành một con số (theo thứ tự từ 0 tăng dần lên 1 đơn vị). Sau đó trả về list đã được mã hóa

    Ví dụ:

    ```
    Cho L = ["đen","vàng","xanh","vàng","xanh","đỏ","hồng"]

    Xây dựng dictionary mã hóa:
    D = {'đen':0,'vàng':1,'xanh':2,'đỏ':3,'hồng':4}

    Trả về List mã hóa:
    L_mahoa = [0, 1, 2, 1, 2, 3, 4]
    ```

````python
def encode_list(L):
    D = {}
    L_mahoa = []
    index = 0
    for item in L:
        if item not in D:
            D[item] = index
            index += 1
        L_mahoa.append(D[item])
    return L_mahoa
L = ["đen","vàng","xanh","vàng","xanh","đỏ","hồng"]
print(encode_list(L))       
````

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
import numpy as np
arr = np.random.randint(0, 1001, size=100)

print("Array:", arr)

# 10 phần tử đầu tiên
print("\n10 phần tử đầu tiên:")
print(arr[:10])

# 5 phần tử cuối cùng
print("\n5 phần tử cuối cùng:")
print(arr[-5:])

# Các phần tử chia hết cho 5 và lớn hơn 100
print("\nCác phần tử chia hết cho 5 và > 100:")
print(arr[(arr % 5 == 0) & (arr > 100)])
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
import numpy as np

# Phát sinh mảng arr gồm 100 số nguyên ngẫu nhiên trong khoảng [0, 1000]
arr = np.random.randint(0, 1001, size=100)

print("Mảng arr:", arr)

# 1. Tính tổng mảng
tong = arr.sum()

# 2. Giá trị trung bình
mean = arr.mean()

# 3. Giá trị trung vị
median = np.median(arr)

# 4. Q1, Q2, Q3
Q1 = np.percentile(arr, 25)   # tứ phân vị thứ nhất
Q2 = np.percentile(arr, 50)   # trung vị
Q3 = np.percentile(arr, 75)   # tứ phân vị thứ ba

# 5. Phân vị 20
P20 = np.percentile(arr, 20)

# In kết quả
print("\n--- Thống kê ---")
print("Tổng mảng:", tong)
print("Giá trị trung bình:", mean)
print("Giá trị trung vị:", median)
print("Q1:", Q1)
print("Q2 (trung vị):", Q2)
print("Q3:", Q3)
print("Phân vị 20:", P20)
````

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

````python
# This cell contains output. Don't remove!
````

````python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu ngẫu nhiên phân phối chuẩn
data = np.random.normal(size=10000)

# Tạo figure gồm 2 subplot (2 hàng, 1 cột)
fig, axs = plt.subplots(2, 1, figsize=(10, 6), gridspec_kw={'height_ratios': [3, 1]})

# Đặt tiêu đề tổng
fig.suptitle("Đồ thị hàm sin(x) và cos(x)", fontsize=14)

# Vẽ histogram
axs[0].hist(data, bins=60, density=True, color='mediumpurple', alpha=0.7)
axs[0].set_title("Histogram")
axs[0].set_ylabel("Mật độ")

# Vẽ boxplot nằm ngang
axs[1].boxplot(data, vert=False)
axs[1].set_title("Boxplot")

# Hiển thị biểu đồ
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````
